```python
import pandas as pd
import matplotlib.pyplot as plt
from pandas import read_csv
from statsmodels.tsa.arima_model import ARIMA
from statsmodels.tsa.arima_model import ARIMAResults

%matplotlib inline
```

```python
btc_model1 = ARIMAResults.load('btc_model1.pkl')
```

```python
btc_model1.summary()
```

```python
eth_model1 = ARIMAResults.load('eth_model1.pkl')
```

```python
doge_model1 = ARIMAResults.load('doge_model1.pkl')
```

```python
ltc_model1 = ARIMAResults.load('ltc_model1.pkl')
```

```python
plots1 = (
    pd.DataFrame(btc_model1.forecast(steps=30)[0]).plot(title="BTC Futures Forecast"),
    pd.DataFrame(eth_model1.forecast(steps=30)[0]).plot(title="ETH Futures Forecast"),
    pd.DataFrame(doge_model1.forecast(steps=30)[0]).plot(title="DOGE Futures Forecast"),
    pd.DataFrame(ltc_model1.forecast(steps=30)[0]).plot(title="LTC Futures Forecast")
)
```
